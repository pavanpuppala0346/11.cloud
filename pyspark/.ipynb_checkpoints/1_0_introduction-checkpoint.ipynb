#### What is pyspark?
- pyspark is an interface for Apache Spark in Python.
- it not only allows you to write spark applications using Python API's, but also provides the PySpark shell for interactively analyzing your data in a distributed environment
- PySpark supports most of Spark's features sucha s Spark SQL, DataFrame, Streaming, MLlib and Spark Core

#### 3. Read single or multiple CSV file in to Dataframe using PySpark (csv('path') or .format('csv').load('path')), display(df)

#### csv file path
- open data --> to get the data files system in databricks
- go to admin console  --> workspace settings --> DBFS File Browser(enable it)
- data --> DBFS --> Files Store --> new folder(data) --> upload csv files
- to copy path of csv files  --> right click on data folder --> copy path

```python
from IPython.display import Image
Image("E:/code/11.cloud/pyspark/img/sp4.png")
```

- df = spark.read.csv(path='dbfs:/FileStore/data/employees1.csv', header=True, sep='|')
- display(df)
- df.printSchema()
- help(StrcutType())

#### to read multiple csv files in different folders
- df = spark.read.csv(path=['dbfs:/FileStore/data1/employees1.csv','dbfs:/FileStore/data2/employees2.csv'], header=True)

#### to read multiple csv files in same folder
- df = spark.read.csv(path='dbfs:/FileStore/data', header=True)

#### Note: schema must be same

- from pyspark.sql.types import *
- schema = StructType().add(field='id',data_type=IntegerType()).add(field='name',data_type=StringType())
- df = spark.read.csv(path='dbfs:/FileStore/data', schema=schema, header=True)
- display(df)
