### [Key Vault](https://vcloud-lab.com/entries/microsoft-azure/create-key-vault-and-secrets-with-access-policies-in-microsoft-azure) 
Azure Key Vault is a cloud service in Microsoft Azure that provides a secure location to store and manage secrets, keys, and certificates, protecting sensitive data like passwords, API keys, and cryptographic keys used by cloud applications and services. It offers centralized management, improved security through Hardware Security Modules (HSMs), and helps to reduce latency and improve the performance of cloud applications by quickly scaling to meet demand.
- it is a centralized sensitive information management vault service
- we can store SSH keys, Secrets, passwords, certificates like database details
- keys: to connect to linux machine we use SSH key  (two types: RSA, EC )
- Secrets : if it is password we can define it in secrets
- Note: for every created object(key, secret, certificate) it generates a **universal resource key identifier**
